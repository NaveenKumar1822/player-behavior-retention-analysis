### Data Cleaning

```python
# Create a working deep copy for data preprocessing
df_clean = df.copy()
```

```python
# Drop metadata and low-variance columns to optimize feature space
df_clean = df_clean.drop(columns=['slug', 'background_image', 'clip', 'short_screenshots', 'saturated_color', 'dominant_color', 'user_game', 'esrb_rating', 'community_rating', 'esrb_rating.id', 'esrb_rating.slug'])
```

```python
# Quantify missing values across remaining features
df_clean.isnull().sum()
```

```python
# Remove records lacking critical release date information
df_clean = df_clean.dropna(subset=['released'])
```

```python
# Impute missing engagement status counts with zero
status_cols = ['added_by_status.beaten', 'added_by_status.toplay', 'added_by_status.dropped', 'added_by_status.playing']

for col in status_cols:
    df_clean[col] = df_clean[col].fillna(0)
```

```python
# Handle missing categorical values by introducing an explicit category
df_clean['esrb_rating.name'] = df['esrb_rating.name'].fillna('Not Rated')
```

### Correct the data types

```python
# Standardize engagement status metrics to integer types
for col in status_cols:
    df_clean[col] = df_clean[col].astype('int')
```

```python
# Convert release date field to standard datetime format
df_clean['released'] = pd.to_datetime(df_clean['released'])
```

### Sanity Check

```python
# Verify data types and schema integrity post-cleaning
df_clean.info()
```

```python
# Confirm all missing values have been successfully addressed
df_clean.isnull().sum()
```

### Feature Engineering

```python
# Define baseline player count (total library additions) for rate derivation
denominator = df_clean['added']
```

```python
# Compute behavioral rates reflecting player completion, abandonment, and active engagement
df_clean['completion_rate'] = (df_clean['added_by_status.beaten'] / denominator) * 100
df_clean['drop_rate'] = (df_clean['added_by_status.dropped'] / denominator) * 100
df_clean['engagement_rate'] = (df_clean['added_by_status.playing'] / denominator) * 100
```

```python
# Standardize rate metrics to two decimal precision
cols_to_round = ['completion_rate', 'drop_rate', 'engagement_rate']
df_clean[cols_to_round] = df_clean[cols_to_round].round(2)
```

```python
# Impute zero for any NaN values resulting from division by zero
df_clean[cols_to_round] = df_clean[cols_to_round].fillna(0)
```

```python
# Filter dataset to establish a high-volume sample (>1,000 additions) to reduce statistical noise
df_popular = df_clean[df_clean['added'] > 1000].copy()
```

```python
# Identify the top 3 titles with the highest abandonment rates
top_dropped = df_popular.sort_values(by='drop_rate', ascending=False).head(3)
```

```python
# Display key behavioral performance metrics for the most dropped games
display(top_dropped[['name', 'added', 'drop_rate', 'completion_rate', 'engagement_rate']])
```

### Exploratory Data Analysis

```python
# Establish uniform professional theme for data visualizations
sns.set_theme(style="whitegrid")
```

#### 1. Do most players actually finish the games they buy?

```python
# Plot the distribution of player completion rates across the popular games cohort
plt.figure(figsize=(19,6))
sns.histplot(data=df_popular, x='completion_rate', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Player Completion Rates', fontsize=14, fontweight='bold')
plt.xlabel('Completion Rate (%)', fontsize=12)
plt.ylabel('Number of Games', fontsize=12)
plt.show()

plt.savefig('Chart1')
```

```python
# Calculate the sample average completion rate
np.mean(df_popular['completion_rate'])
```

#### 2. Does a higher rating prevent players from dropping a game?

```python
# Analyze the relationship between overall review rating and game drop rates
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_popular, x='rating', y='drop_rate', alpha=0.6, color='coral')
plt.title('Drop Rate vs. Average Game Rating', fontsize=14, fontweight='bold')
plt.xlabel('Game Rating (Max 5.0)', fontsize=12)
plt.ylabel('Drop Rate (%)', fontsize=12)
plt.show()

plt.savefig('Chart2')
```

#### 3. Which games have the highest active player base right now?

```python
# Filter and extract the top titles by active engagement rate
top_1_engaged = df_popular.sort_values(by='engagement_rate', ascending=False).head()
```

```python
# Visualize current active player engagement across top titles
plt.figure(figsize=(12, 6))
sns.barplot(data=top_1_engaged, x='name', y='engagement_rate', hue='name', palette='viridis', legend=False)
plt.title('Top 10 Games by Current Engagement Rate', fontsize=14, fontweight='bold')
plt.xlabel('Game Title', fontsize=12)
plt.ylabel('Engagement Rate (% of Owners Playing)', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.savefig('Chart3')
```
